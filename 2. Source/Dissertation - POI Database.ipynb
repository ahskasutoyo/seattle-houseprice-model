# Processing Data

```python
# Normalize the 'types' column which contains lists of types
df['types'] = df['types'].apply(eval)  # Make sure the types are in list format
types_expanded = df.explode('types')  # Create a new row for each type per POI

# Count the occurrences of each type
type_counts = types_expanded['types'].value_counts()

print(type_counts)
```

```python
# Save the combined DataFrame to a new CSV file
types_expanded.to_csv('POI_database.csv', index=False)

print("Combined dataset saved as 'POI_database.csv'")
```

# Remove Duplicate Types

```python
# Load the data
file_path = 'C:/Users/ahska/Documents/Ahska/UCL/3 - Final Project (Dissertation)/Dissertation 2/POI Data/poi_database(2) v2.csv'  # Update this to the path of your CSV file
data = pd.read_csv(file_path)

# Define your priority list of types
priority_list = ['parking', 'police', 'restaurant', 'fire_station', 'train_station', 'convenience_store', 'cafe', 
                 'doctor', 'secondary_school', 'pharmacy', 'light_rail_station', 'subway_station', 'primary_school', 
                 'bar', 'drugstore', 'hospital', 'night_club', 'transit_station', 'bus_station', 'park', 
                 'grocery_or_supermarket', 'shopping_mall', 'department_store', 'movie_theater', 'university', 
                 'ferry_terminal'] # Update as needed

# Create a priority column based on the types, where lower numbers have higher priority
data['priority'] = data['types'].apply(lambda x: priority_list.index(x) if x in priority_list else len(priority_list))

# Sort the data by place_id and priority (lower numbers first)
data_sorted = data.sort_values(by=['place_id', 'priority'])

# Drop duplicates, keeping the first entry per place_id (highest priority)
data_unique = data_sorted.drop_duplicates(subset='place_id', keep='first')

# Optionally, count the original and unique entries to see how many duplicates were removed
original_count = data.shape[0]
unique_count = data_unique.shape[0]
duplicate_count = original_count - unique_count

print(f"Original Count: {original_count}, Unique Count: {unique_count}, Duplicates Removed: {duplicate_count}")

# Save the processed data if necessary
data_unique.to_csv('processed_data.csv', index=False)
```
