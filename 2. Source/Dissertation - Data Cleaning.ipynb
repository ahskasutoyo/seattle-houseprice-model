```python
# Convert 'SOLD DATE' to datetime format
data['SOLD DATE'] = pd.to_datetime(data['SOLD DATE'], errors='coerce')
```

```python
# Count the number of houses for each property type meeting either condition
property_type = data['PROPERTY TYPE'].value_counts()

# Display results
print("\nNumber of houses for each property type with (BEDS or BATHS):")
print(property_type)
```

```python
# Display counts of Houses with 0 and blank (NaN) 'BEDS'
beds_zero = data[data['BEDS'] == 0].shape[0]
beds_nan = data['BEDS'].isna().sum()

# Display counts of Houses with 0, 0.5, and blank (NaN) 'BATHS'
baths_zero = data[data['BATHS'] == 0].shape[0]
baths_half = data[data['BATHS'] == 0.5].shape[0]
baths_nan = data['BATHS'].isna().sum()

print("Number of Houses with 0 Beds:", beds_zero)
print("Number of Houses with blank Beds:", beds_nan)
print("Number of Houses with 0 Baths:", baths_zero)
print("Number of Houses with 0.5 Baths:", baths_half)
print("Number of Houses with blank Baths:", baths_nan)
```

```python
# Identify houses with 0 or NaN 'BEDS'
beds_zero_or_nan = data[(data['BEDS'] == 0) | (data['BEDS'].isna())]

# Identify houses with 0, 0.5, or NaN 'BATHS'
baths_zero_half_or_nan = data[(data['BATHS'] == 0) | (data['BATHS'] == 0.5) | (data['BATHS'].isna())]

# Combine the filters for BEDS and BATHS
combined_filter = data[(data['BEDS'] == 0) | (data['BEDS'].isna()) |
                       (data['BATHS'] == 0) | (data['BATHS'] == 0.5) | (data['BATHS'].isna())]

# Count the number of houses for each property type meeting the BEDS condition
beds_counts = beds_zero_or_nan['PROPERTY TYPE'].value_counts()

# Count the number of houses for each property type meeting the BATHS condition
baths_counts = baths_zero_half_or_nan['PROPERTY TYPE'].value_counts()

# Count the number of houses for each property type meeting either condition
combined_counts = combined_filter['PROPERTY TYPE'].value_counts()

# Display results
print("Number of houses for each property type with 0 or NaN BEDS:")
print(beds_counts)
print("\nNumber of houses for each property type with 0, 0.5, or NaN BATHS:")
print(baths_counts)
print("\nNumber of houses for each property type with either condition (BEDS or BATHS):")
print(combined_counts)
```
